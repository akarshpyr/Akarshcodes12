```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x = df1['Length of Membership'].values.reshape(-1,1)
y = df1['Yearly Amount Spent'].values.reshape(-1,1)
```

```python
x
```

```python
x.shape
```

```python
x_train, x_test, y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=12)
```

```python
x_train
```

```python
x_train.shape
```

```python
np.arange(1,12)
```

```python
model_lr = LinearRegression()
```

```python
model_lr.fit(x_train,y_train)
```

```python
y_pred = model_lr.predict(x_test)
```

```python
y_pred[:5]
```

```python
from sklearn.metrics import r2_score
```

```python
R2 = r2_score(y_test,y_pred)
```

```python
R2
```

```python
import statsmodels.api as sm
```

```python
X = df1['Length of Membership']
Y = df1['Yearly Amount Spent']
```

```python
X_train, X_test, Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state=12)
```

```python
x_reg = sm.add_constant(X_train)
```

```python
x_reg
```

```python
model = sm.OLS(endog=Y_train,exog=x_reg[['const','Length of Membership']])
```

```python
res = model.fit()
```

```python
print(res.summary())
```
